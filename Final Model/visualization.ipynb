```python
prstd, iv_l, iv_u = wls_prediction_std(res)
fig, ax = plt.subplots(figsize=(10,3))
ax.scatter(X,y,c='b',s=12,alpha=0.6)
ax.plot(X,res.fittedvalues,'r')
ax.plot(X,iv_l,'--',color='grey',linewidth=1)
ax.plot(X,iv_u,'--',color='grey',linewidth=1)
# plt.fill_between(X,iv_l,iv_u,color='blue',alpha=0.1)
plt.xlim(0.5,10.5)
plt.ylim(-350,1150)
ax.legend(['fitted','lower bound','upper bound'],prop={'size': 18})
plt.savefig('lin_original.pdf')
```
